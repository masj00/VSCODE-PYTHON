# Sets
## Valgfagseksempler

```python
python          = {"Alice", "Bob", "Charlie", "Diana"}
it_sikkerhed    = {"Bob", "Charlie", "Eva", "Frank"}
ai_agenter      = {"Alice", "Eva", "Georg", "Helen"}
```

### Hvem tager Python ELLER IT-sikkerhed? Union


```python
print("Python eller IT-sikkerhed:", python | it_sikkerhed)

python.union(it_sikkerhed)
```

### Hvem tager BÅDE Python OG IT-sikkerhed?

```python
print("Begge kurser:", python & it_sikkerhed)

python.intersection(it_sikkerhed)
```

### Hvem tager Python men IKKE IT-sikkerhed?

```python
print("Kun Python:", python - it_sikkerhed)

python.difference(it_sikkerhed)
```

## Hvem tager IT-sikkerhed men IKKE Python?

```python
print("Kun IT-sikkerhed:", it_sikkerhed - python)
it_sikkerhed.difference(python)
```

## Hvem tager præcis EN af disse?

```python
print("Præcis en:", python ^ it_sikkerhed)
python.symmetric_difference(it_sikkerhed)
```

### Hvem tager Python eller IT-sikkerhed, men ikke AI-agenter?

```python
print("Python/IT-sikkerhed men ikke AI-agenter:", (python | it_sikkerhed) - ai_agenter)
(python | it_sikkerhed).difference(ai_agenter)
```

# Dictionaries

### Basic students-karakter dictionary

```python
grades = {
    "Alice": 12,
    "Bob": 7,
    "Charlie": 10,
    "Diana": 4
}
```

```python
# Hurtig "lookup"
grades['Bob']
```

```python
# Keys skal være 'unique' og 'imutable'
# * ikke 2 Alice
# * Strings, int, tubple - ikke lists 

```

```python
# Values kan være alt
# * Lists, tuples, strings. int ...
```

### Metoder

```python
# alle keys
print("Alle studerende:", list(grades.keys()))

# alle values  
print("Alle karakterer:", list(grades.values()))

# key-value pairs (her har vi en liste med tuples)
print("Alle key / value par:", list(grades.items()))

# "Safe" access
print("xxx's karakter:", grades.get("xxx", "Ingen karakter"))

# Adding/updating
grades["Eva"] = 10
grades["Alice"] = 12  # opdaterer eksisterende
```

### Udvidet kode

```python
#

students = {
    "Alice": {"age": 22, "courses": ["Python", "Math"], "grade": 12},
    "Bob": {"age": 21, "courses": ["Python", "Physics"], "grade": 7},
    "Charlie": {"age": 23, "courses": ["Math", "Physics"], "grade": 10}
}
```

```python
# Accessing nested data
students['Alice']['age']
students['Alice']['courses'][0]

# Iterating through dictionary
for name, info in students.items():
    print(f"{name} is {info['age']} years old and enrolled in {', '.join(info['courses'])}.")

# Finding students in specific courses
python_students = [name for name, info in students.items() if "Python" in info['courses']]
print("Students enrolled in Python:", python_students)

# Dictionary comprehension version

```

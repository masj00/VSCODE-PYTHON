```python
liste = [2,1,3,4,5,6]
print(liste)
```

```python
tuple = (1, 'Volvo', 2.34, 700)
print(tuple)
```

```python
#Hvordan for man fat i et element fra en liste
print(liste[1])
#og tuple
print(tuple[1])
```

```python
#Iterer
names = ['Hej', 'Marco', 'is', 'learning', 'Python']

#for each
for element in liste:
    print(element)

#For loop
for x in range(1,5):
    print(x)

#enumerate putter et index på hvert element i en liste
for index, value in enumerate(names):
    print(index, value)
```

```python
#Slicing
# Første til og med tredje element (indeks 0 til 2)
sub1 = liste[0:3]
print(sub1)  # [1, 2, 3]

# Tredje til femte element (indeks 2 til 4)
sub2 = liste[2:5]
print(sub2)  # [3, 4, 5]

# Fra midten af listen til slut
midten = len(liste) // 2
sub3 = liste[midten:]
print(sub3)  # [4, 5, 6]

# Fra start til slut, men kun hver anden element
sub4 = liste[::2]
print(sub4)  # [1, 3, 5]
```

```python
#Sortering
liste.sort()
print(liste)

# Denne redigere den oprindelig liste og laver den om
names.sort(key=len, reverse=True)
print(names)

#Dette er den bedste (den retunere en ny liste)
sorted(names)

def secondLast(x):
    return x[-2]

sorted(names, key=secondLast)
```

```python
#List Comprehension
listcomp = [x for x in range(1, 6)]
print(listcomp)  

listcomp1 = [i for i in range(1,11) if (i != 5) ^ (i != 8)]
print(listcomp1)
listcomp2 = [i for i in range(1, 11) if i != 5 and i != 8]
print(listcomp2) 
```

```python
# abc = [x for x in range(65,90)]
# print(abc)

abc = [chr(x) for x in range(65, 91)]
print(abc) 

#ord('A')=65 chr(65)=A

# Store bogstaver A-Z, udelader dem med Unicode 70, 75, 80, 85
abcUdenFire = [chr(x) for x in range(65, 91) if x not in (70, 75, 80, 85)]
print(abcUdenFire)

# Store bogstaver A-Z, udelader hvert andet bogstav mellem F (70) & O (79)
abcUdenHvertAndet = [
    chr(x) for x in range(65, 91) if not (70 <= x <= 79 and (x - 70) % 2 == 0)
]
print(abcUdenHvertAndet)

```

```python
ligetal = [i  for i in range(1,21) if i % 2 == 0]
#if i % 2 == 0 betyder: kun tag tal, hvor resten ved division med 2 er 0 
# (altså lige tal). fx 3 % 2 = 1 så det er ikke 0 derfor tæller den ikke med
print(ligetal)

kvadrattal = [i**2 for i in range(1,11)]
# i**2 betyder "i opløftet i anden" (altså i gange i).
print(kvadrattal)

#Lav en liste af alle vokaler i en given streng.
findvokaler = "Hej jeg spiser et æble"
vokaler = [x for x in findvokaler if x.lower() in 'aeiouæøå']
#x.lower() in 'aeiouæøå' betyder:
#Tjek om det lille bogstav af x findes i strengen 'aeiouæøå'.
#Hvis det gør, så er x en vokal, og den bliver taget med i listen.

print(vokaler)
# char.lower() gør at store bogstaver også tæller med


```

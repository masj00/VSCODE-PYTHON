```python
def has_pair_with_sum(data, target_sum):
    low = 0
    hi = len(data) - 1

    while low < hi:
        s = data[low] + data[hi]
        if s == target_sum:
            return True
        elif s < target_sum:
            low += 1
        else:
            hi -= 1
            return False
        

print(has_pair_with_sum([1,2,3,9], 8))
print(has_pair_with_sum([1,2,4,4], 8))
```

```python
def has_pair_with_sum(data, target_sum):
    seen = set()
    return any(target_sum - x in seen or seen.add(x) for x in data)

print(has_pair_with_sum([1,3,3,9], 8))
print(has_pair_with_sum([1,2,4,4], 8))
```
